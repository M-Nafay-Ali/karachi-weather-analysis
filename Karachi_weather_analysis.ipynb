```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/predicted_temperatures.csv")
print(df)

print("Now displaying information")
print(df.info())

print("Now understanding data")
print(df.head())
```

As we have two different algorithms(ARIMAX and LassoRegressor)
We need to find out which of them is winner and we do that by calculate Mean-Absolute-Error(MAE)
As we first filter out those two categories and then calculate their error  with function "mean_absolute_error()" with both types actual temperature and their predicted temperature as they are given in data set to know how much is there an error in them

```python
from sklearn.metrics import mean_absolute_error

# 1. Filter data for each algorithm
arimax_data = df[df['Algorithm'] == 'ARIMAX']
lasso_data = df[df['Algorithm'] == 'LassoRegressor']

# 2. Calculate Error (Lower is better)
mae_arimax = mean_absolute_error(arimax_data['Actual'], arimax_data['Predicted_Temperature'])
mae_lasso = mean_absolute_error(lasso_data['Actual'], lasso_data['Predicted_Temperature'])

print(f"ARIMAX Error: {mae_arimax:.2f}°C")
print(f"Lasso Error: {mae_lasso:.2f}°C")
```

To understand which model is better for the Karachi climate, we are using an Interactive Time-Series Dashboard.

First we check that if date is in correct format
Then we create lively graph with the help of Plotly library
Then we add "Actual" temperature and Date as x and y axis
Then after that we create another graph of ARIMAX with respect to Date and predicted temperature and same goes for LassoRegressor and
In last we add interactive  features such as title,colors,template as we choose dark,add hovermode and then set x axis and y axis

```python
import plotly.graph_objects as go

# 1. Ensure Date is correctly formatted for the range slider
df['Datetime'] = pd.to_datetime(df['Datetime'])

# 2. Create the "Lively" Figure
fig = go.Figure()

# Add Actual Temperature (The truth)
fig.add_trace(go.Scatter(
    x=df['Datetime'], y=df['Actual'],
    name='Actual Karachi Temp',
    line=dict(color='#00CC96', width=3) # Bright Green
))

# Add ARIMAX Prediction (Filtered from your Algorithm column)
arimax_df = df[df['Algorithm'] == 'ARIMAX']
fig.add_trace(go.Scatter(
    x=arimax_df['Datetime'], y=arimax_df['Predicted_Temperature'],
    name='ARIMAX Prediction',
    line=dict(color='#636EFA', dash='dash') # Blue dashed
))

# Add Lasso Prediction (Filtered from your Algorithm column)
lasso_df = df[df['Algorithm'] == 'LassoRegressor']
fig.add_trace(go.Scatter(
    x=lasso_df['Datetime'], y=lasso_df['Predicted_Temperature'],
    name='Lasso Prediction',
    line=dict(color='#EF553B', dash='dot') # Red dotted
))

# 3. Add the Interactive "Lively" Features
fig.update_layout(
    title='Interactive Karachi Weather: Model Comparison',
    template='plotly_dark',
    hovermode='x unified', # Shows all values at once when you hover
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=3, label="3m", step="month", stepmode="backward"),
                dict(step="all", label="All Time")
            ]),
            font=dict(color="black")
        ),
        rangeslider=dict(visible=True), # Adds the bottom slider for zooming
        type="date"),
    yaxis_title="Temperature (°C)")

fig.show()
```

Since ARIMAX has obtained lower error it is Winner and which make it more suitable for short-term daily predictions
As ARIMAX was best in capturing Seasonality of Karachi
